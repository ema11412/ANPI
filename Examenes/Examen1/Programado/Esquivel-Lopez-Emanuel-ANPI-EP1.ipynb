```sos
clc; clear
pkg load dataframe
pkg load symbolic

function [df] = derivar(f,x)
    %Utilice esta funcion para realizar los calulos de derivadas
    
    aux = strcat('@(x)',f); 
    f1 = str2func(aux);
    h=0.000000001;
    df = (f1(x + h) - f1(x)) / h;
endfunction

function [x,k]=newton(f,x0,tol,iterMax)  
    %Esta funci�n aproxima la soluci�n de la ecuaci�n f(x)=0, utilizando el m�todo de la newton
    %
    %Sintaxis:  newton(f,x0,tol,iterMax)
    % 
    %Par�metros Iniciales: 
    %Entra una funcion definida arriba, por el parametro de entrada x0 
    %utilizads en la formula recursiva          
    %            
    %            
    %Par�metros de Salida:                           
    %Muestra una grafica con los valores de el error 
    %Muestra la ultima iteracion y su valor calculado           
    %  
    aux = strcat('@(x)',f); % @x + funci�n
    f1 = str2func(aux); % funci�n string a ecuaci�n

    x=[]; k=[]; error=[];

    k=0;
    error=tol+1;
    e=[];
        while and(error>tol,k<iterMax)
            k=k+1;
            if ((derivar(f,x0) == 0))
              break;  
            endif
            x= x0-(f1(x0)/(derivar(f,x0)));
            error=abs(f1(x));
            e=[e error];
            x0=x;
        endwhile
        %plot(1:k,e)
end

function [x,k]=steffensen(f,x0,tol,iterMax)  
    %Esta funci�n aproxima la soluci�n de la ecuaci�n f(x)=0, utilizando el m�todo de la newton
    %
    %Sintaxis:  steffensen(f,x0,tol,iterMax)
    % 
    %Par�metros Iniciales: 
    %Entra una funcion definida arriba, por el parametro de entrada x0 
    %utilizads en la formula recursiva          
    %            
    %            
    %Par�metros de Salida:                           
    %Muestra una grafica con los valores de el error 
    %Muestra la ultima iteracion y su valor calculado           
    %  
    aux = strcat('@(x)',f); % @x + funci�n
    f1 = str2func(aux); % funci�n string a ecuaci�n
    x=[]; k=[]; error=[];
    k=0;
    error=tol+1;
    e=[];
        while and(error>tol,k<iterMax)
            if ((f1(x0+f1(x0))-f1(x0)) == 0)
              break; 
            endif
            k=k+1;
            x= x0-((f1(x0))^2/(f1(x0+f1(x0))-f1(x0)));
            error=abs(f1(x));
            e=[e error];
            x0=x;
        endwhile
        %plot(1:k,e)
end

function [x,k]=NewtonSteff(f,x0,tol,iterMax)  
    %Esta funci�n aproxima la soluci�n de la ecuaci�n f(x)=0, utilizando el m�todo de la newton
    %
    %Sintaxis:  NewtonSteff(f,x0,tol,iterMax)
    % 
    %Par�metros Iniciales: 
    %Entra una funcion definida arriba, por el parametro de entrada x0 
    %utilizads en la formula recursiva          
    %            
    %            
    %Par�metros de Salida:                           
    %Muestra una grafica con los valores de el error 
    %Muestra la ultima iteracion y su valor calculado           
    %  
    aux = strcat('@(x)',f); % @x + funci�n
    f1 = str2func(aux); % funci�n string a ecuaci�n
    x=[]; k=[]; error=[];
    k=0;
    error=tol+1;
    e=[];
        while and(error>tol,k<iterMax)
            k=k+1;
            if ((derivar(f,x0)) == 0)
              break; 
            endif            
            xl=x0-(f1(x0)/(derivar(f,x0)));
            
            if ((derivar(f,x0))*(f1(x0)-f1(xl)) == 0)
              break; 
            endif            
            x=x0-((f1(x0))^2/((derivar(f,x0))*(f1(x0)-f1(xl))));
            error=abs(f1(x));
            e=[e error];
            x0=x;
        endwhile
        %plot(1:k,e)
end
x01=2;
x02=1;
x03=15;
%primera prueba atan(x)
[x11,k11]=newton('atan(x)',x01,10^-10,10);
[x21,k21]=steffensen('atan(x)',x01,10^-10,10);
[x31,k31]=NewtonSteff('atan(x)',x01,10^-10,10);
%segunda prueba sin(x)-x/2
[x12,k12]=newton('sin(x)-x/2',x01,10^-10,10);
[x22,k22]=steffensen('sin(x)-x/2',x01,10^-10,10);
[x33,k32]=NewtonSteff('sin(x)-x/2',x01,10^-10,10);
%tercera prueba 10*x*exp(-x^2)-1
[x13,k13]=newton('10*x*exp(-x^2)-1',x02,10^-10,10);
[x23,k23]=steffensen('10*x*exp(-x^2)-1',x02,10^-10,10);
[x33,k33]=NewtonSteff('10*x*exp(-x^2)-1',x02,10^-10,10);
%cuarta prueba x^6-36*x^5+450*x^4-2400*x^3+5400*x^2-4320*x+720
[x14,k14]=newton('x^6-36*x^5+450*x^4-2400*x^3+5400*x^2-4320*x+720',x03,10^-10,10);
[x24,k24]=steffensen('x^6-36*x^5+450*x^4-2400*x^3+5400*x^2-4320*x+720',x03,10^-10,10);
[x34,k34]=NewtonSteff('x^6-36*x^5+450*x^4-2400*x^3+5400*x^2-4320*x+720',x03,10^-10,10);
%quinta prueba x*log10(x)-1.2
[x15,k15]=newton('x*log10(x)-1.2',x01,10^-10,10);
[x25,k25]=steffensen('x*log10(x)-1.2',x01,10^-10,10);
[x35,k35]=NewtonSteff('x*log10(x)-1.2',x01,10^-10,10);

C = {"f(x)"          , "x0", "Raiz", "Iteracion NM", "Iteracion SM", "Iteracion NSM"; 
     "atan(x)"       , x01 , 0.00000000000000 , k11,k21,k31; 
     "sin(x)-x/2"    , x01 , 1.89549426703398 , k12,k22,k32;
     "10xexp(-x^2)-1", x02 , 1.67963061042845 , k13,k23,k33;
     "x^6-36x^5+.."  , x03 , 15.98287398060170, k14,k24,k34;
     "xlog10(x)-1.2" , x01 , 2.74064609597369 , k15,k25,k35};
dataframe (C)
```

```sos
import numpy as np  

#Debido a que en el enunciado se desea SOLO resolver la matriz A 100x100 y un vector d de 100 elementos
#No posee valores iniciales, son generados dentro de la funcion, el unico valor es el orden de la matriz A
def Thomas(tam): 
    #recibe de entrada el valor del orden de la matriz A el cual debe ser  0<tam<100
    if tam>100 or type(tam) != int or tam<=0:
        return "error, orden de matriz invalido"
    #se procede a llenar la parte inferior de la diagonal
    a=llenarVector("subD1",tam)
    #se procede a llenar la diagonal de la matriz
    b=llenarVector("diagonal",tam)
    #se procede a llenar la parte superor de la diagonal
    c=llenarVector("subD2",tam)
    #se procede a llenar el vector d
    d=llenarVector("d",tam)
    n=np.shape(a)[0]  
    
    #se llenan de ceros los vectores q, p, x
    p = np.zeros(n) 
    q = np.zeros(n) 
    x  = np.zeros(n)  
    
    #se realiza la primera iteracion de la formula
    p[0]=c[0]/b[0]  
    for i in range(1,n-1,1):  
       p[i]=c[i]/(b[i]-a[i]*p[i-1])  
    
    #se realiza la segunda iteracion de la formula
    q[0]=d[0]/b[0]  
    for i in range(1,n,1):  
       q[i]=(d[i]-a[i]*q[i-1])/(b[i]-a[i]*p[i-1])  

    #se realiza la ultima iteracion para encontrar el valor de x
    x[n-1]=q[n-1]  
    for i in range(n-2,-1,-1):  
       x[i]=q[i]-p[i]*x[i+1]  
 
    return x  

#Esta funcion se encarga de llenar los elementos a, b, c, d, los cuales fueron especificados
def llenarVector(Valor,largo):
    #recibe de entrada un valor, el cual sera dependiendo el vector se quiera llenar
    #subD1 -> a
    #diagonal -> b
    #subD2 -> c
    #d -> d
    vec = []
    if Valor == "diagonal":
        k=0
        while k<largo:
            vec +=[5]
            k+=1
        return vec
    if Valor == "subD2":
        k=0
        while k<largo:
            if k==largo-1:
                vec+=[0]
                k+=1
            else:
                vec +=[1]
                k+=1
        return vec
    if Valor == "subD1":
        k=0
        while k<largo:
            if k==0:
                vec+=[0]
                k+=1
            else:
                vec +=[1]
                k+=1
        return vec
    else:
        k=0
        while k<largo:
            if k==0 or k==largo-1:
                vec+=[-12]
                k+=1
            else:
                vec+=[-14]
                k+=1
        return vec
#se llama con la base del problema, una matriz cuadrada de orden 100
Thomas(100)
```
